# Data Cleaning

```python
# Create a new dataframe as a copy of our original dataframe
cleaned_df = df.copy()
```

```python
# Clean the first column, Invoice

# Adjust data type to string so that we can apply regex as there are specific patterns that we want to filter out, aka those with letter "C" or "A"
cleaned_df["Invoice"] = cleaned_df["Invoice"].astype("str")

# Create a new variable "mask" as a filter expression
mask = (
    cleaned_df["Invoice"].str.match("^\\d{6}$") == True
)

# Apply filter "mask" to get Invoice with only 6-digit code -- here we got 515,252 records
cleaned_df = cleaned_df[mask]
cleaned_df
```

```python
# Adjust a variable "mask" as a filter expression for StockCode column
cleaned_df["StockCode"] = cleaned_df["StockCode"].astype("str")

mask = (
    (cleaned_df["StockCode"].str.match("^\\d{5}$") == True)
    | (cleaned_df["StockCode"].str.match("^\\d{5}[a-zA-Z]+$") == True) # Matches StockCode that starts with one-or-more upper-or lower case letters and also ends with it
    | (cleaned_df["StockCode"].str.match("^PADS$") == True)
)   

cleaned_df = cleaned_df[mask]
cleaned_df
```

```python
# "Description" Column shouldn't have any issue
# "Quantity" column may contain negative values, and from our EDA, negative values tend to belong to Null "CustomerID".

# Remove Null CustomerID
cleaned_df.dropna(subset=["Customer ID"], inplace = True)
cleaned_df.describe()
```

```python
# "Price" Column appears to have 0 as minimum, which may impact our customer analysis. Let's check how many instances of 0 price and remove them
len(cleaned_df[cleaned_df["Price"] == 0])
```

```python
# Only 28 instances of 0 price so it is okay to drop these records
cleaned_df = cleaned_df[cleaned_df["Price"] > 0]

cleaned_df.describe()
```

```python
# It appears that there are instances whose prices are around 0. Let's check them
print(cleaned_df["Price"].min())
print(len(cleaned_df[cleaned_df["Price"] == cleaned_df["Price"].min()]))
# Since these $0.01 records account for only 14 rows, it is okay to include them
```

```python
# Export cleaned dataframe for further customer analysis
cleaned_df.to_excel('cleaned_df.xlsx', index=False)
```

```python
# Data cleaning is completed. Let's check how much data we have left
print(f"Data dropped after cleaning: {1 - len(cleaned_df)/len(df):.2%}")
print(f"Data remain after cleaning: {len(cleaned_df)/len(df):.2%}")
print(f"Rows dropped after cleaning: {df.shape[0] - cleaned_df.shape[0]:,} rows ({1 - len(cleaned_df)/len(df):.2%})")
print(f"Rows remain after cleaning: {cleaned_df.shape[0]:,} rows ({len(cleaned_df)/len(df):.2%})")
```

## Assumptions
Negative Quantity = Invalid Transaction
<li> Assumes negative quantities are errors or irrelevant (e.g., returns or adjustments). </li>
<li> Assumes they shouldn't be included in sales or customer analysis. </li>

Unusual StockCode = Non-Standard Product
<li>Assumes only codes following a strict pattern (e.g., 5-digit or 5-digit + letter, PADS) are valid.</li>
<li>Excludes promotional items, packaging, or services that may use different codes.</li>

Invoices with "C" or "A" = Non-Sales Events
<li>Assumes cancellations ("C") and accounting adjustments ("A") should be excluded from customer analysis.</li>

Null CustomerID = Non-Actionable Record
<li>Assumes records without customer IDs can't be linked to meaningful behavior or segmentation.</li>
